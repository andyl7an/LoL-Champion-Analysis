```python
import numpy as np
import pandas as pd
%matplotlib inline
from matplotlib import pyplot as plt
import seaborn as sns
import glob
import os
from IPython.html.widgets import interact, interactive, fixed
from IPython.html import widgets
import ntpath
```

## Opening my data...

Here, I'm opening my rank and specilization data and appending it to a data frame.

```python
rankInfo = []
with open('otpRankData.txt') as f:
    for line in f:
        line = line.split(',')
        line[1] = line[1].rstrip('\n')
        rankInfo.append(line)
```

```python
df = pd.DataFrame(rankInfo, columns = ('Rank','One Trick Pony'))
df.head()
```

```python
rankOrder = ['BRONZEV','BRONZEIV','BRONZEIII','BRONZEII','BRONZEI',
             'SILVERV','SILVERIV','SILVERIII','SILVERII','SILVERI',
            'GOLDV','GOLDIV','GOLDIII','GOLDII','GOLDI',
            'PLATINUMV','PLATINUMIV','PLATINUMIII','PLATINUMII','PLATINUMI',
            'DIAMONDV','DIAMONDIV','DIAMONDIII','DIAMONDII','DIAMONDI',
             'MASTERI','CHALLENGERI'
            ]
```

This is a quick function to convert my count information to a line graph of the frequency of each rank

```python
def getFreq(input):
    counts = input.value_counts()
    length = len(input)
    array = []
    for i in rankOrder:
        array.append(counts[i] / length)
    temp = range(len(rankOrder))
    plt.plot(temp, array)
```

## The Entire Sample

```python
g = sns.countplot(df['Rank'], order = rankOrder)
g.set_xticklabels(labels = rankOrder, rotation=90);

plt.title('Countplot of Rank');
```

```python
getFreq(df['Rank'])
plt.title('Frequency of Ranks');
plt.xlabel('Ranks')
plt.ylabel('Percentage of Total');
```

Now I have to seperate my sample into specialized and non specialized players

```python
grouped = df.groupby('One Trick Pony')
g = grouped.Rank
notOTP = g.groups['False']
isOTP = g.groups['True']
otpRanks = df['Rank'][isOTP]
nonOtpRanks = df['Rank'][notOTP]
```

## Specialized Players vs Non- Specialized Players

```python
g = sns.countplot(otpRanks, order = rankOrder)
g.set_xticklabels(labels = rankOrder, rotation=90);
plt.title('Count of Specialized Player Ranks');
```

```python
g = sns.countplot(nonOtpRanks, order = rankOrder)
g.set_xticklabels(labels = rankOrder, rotation=90);
plt.title('Count of Non Specialized Player Ranks');
```

```python
getFreq(otpRanks)

getFreq(nonOtpRanks)
plt.title('Frequency of Ranks')
plt.xlabel('Ranks')
plt.ylabel('Percentage of Total');
plt.legend(['Specialized', 'Not Specialized']);
```

So it seems there isn't that much of an effect of specialization on player ranking with the exception of getting from Bronze rank to Silver.

## Correlations in Player Usage of Champions

Now I'll look at the correlation between a player's most played champion and their other ones. Of course, my data is still in the for of "Champion Points" which correlated with their time spent on the champion, so I will need to convert it to a format that makes more sense.

```python
champions = []
path = os.getcwd() + '/champData/'
champions =glob.glob(path+ '*')
champDict= {}

for champ in champions:
    with open(champ) as f:
        c = ntpath.basename(champ)
        for line in f:
            line = line.split(':')
            line[1] = line[1].rstrip('\n')

            champName = line[0]
            champPoints = line[1]

            champPoints = ''.join(x for x in champPoints if x.isdigit())

            if c not in champDict:
                champDict[c] = {}

            if champName not in champDict[c]:
                champDict[c][champName] = float(champPoints)
            else:
                champDict[c][champName] = float(champDict[c][champName]) + float(champPoints)
        
            
```

```python
df = pd.DataFrame.from_dict(champDict)
```

```python
df.head()
```

```python
champColumns = list(df.columns.values)
champIndices = list(df.index.values)
```

This is a helper method that plots the correlation between two champions using the data I've cleaned up in the dataframe

```python
def twoChamp(champA, champB):
    if (champA == champB):
        print('Same Champion, Illegal Argument')
    try:
        a = df[champA][champB]
        b = df[champB][champA]
        plt.scatter(a,b);
        plt.title('Correlation between Champ A and Champ B')
        plt.xlabel('Champ A players on Champ B')
        plt.ylabel('Champ B players on Champ A')
        plt.axis([0, 1.05, 0, 1.05])
    except KeyError:
        print('Not Enough Data')
        
    
```

```python
maxNum = df.max()
i = 0
for col in champColumns:
    df[col] = df[col] / maxNum[i]
    i = i + 1
```

## Some Interactive Stuff

This lets me play around with the champions to see how they correlate with each other. Of course, a champion can't correlate with itself so there won't be any data for those cases.
This is split into champion A and champion B because if a player plays champion A also plays champion B, the reverse is not necessarily true.

```python
interact(twoChamp, champA = champColumns, champB = champIndices);
```

This is a better overview of just one champion, it shows how often players who play this champion would play every other champion in the game.

```python
def champPlayCount(champ):
    plt.figure(figsize = (5,25))
    plt.axhline(0, color='k');
    df.ix[champ].plot.barh();
```

```python
interact(champPlayCount, champ = champColumns);
```
